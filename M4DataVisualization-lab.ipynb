## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
QUERY = """
SELECT ConvertedComp
FROM master
"""

df = pd.read_sql_query(QUERY, conn)

# Plotting the histogram
plt.figure(figsize=(10, 6))
plt.hist(df['ConvertedComp'].dropna(), bins=30, color='blue', edgecolor='black')
plt.title('Histogram of Converted Compensation')
plt.xlabel('Converted Compensation')
plt.ylabel('Frequency')
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here

# Query to get Age data
QUERY = """
SELECT Age
FROM master
"""

df = pd.read_sql_query(QUERY, conn)

# Plotting the box plot
plt.figure(figsize=(10, 6))
sns.boxplot(df['Age'])
plt.title('Box Plot of Age')
plt.xlabel('Age')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here

# your code goes here

# Query to get Age and WorkWeekHrs data
QUERY = """
SELECT Age, WorkWeekHrs
FROM master
"""

df = pd.read_sql_query(QUERY, conn)

# Plotting the scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='WorkWeekHrs', data=df)
plt.title('Scatter Plot of Age vs Work Week Hours')
plt.xlabel('Age')
plt.ylabel('Work Week Hours')
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here

# Query to get Age, WorkWeekHrs, and CodeRevHrs data
QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master
"""

df = pd.read_sql_query(QUERY, conn)

# Plotting the bubble plot
plt.figure(figsize=(10, 6))
plt.scatter(x='WorkWeekHrs', y='CodeRevHrs', s=df['Age'], alpha=0.5, data=df)
plt.title('Bubble Plot of Work Week Hours vs Code Review Hours')
plt.xlabel('Work Week Hours')
plt.ylabel('Code Review Hours')
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# Query to get DatabaseDesireNextYear data
QUERY = """
SELECT DatabaseDesireNextYear, COUNT(*) as count
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY count DESC
LIMIT 5
"""

df = pd.read_sql_query(QUERY, conn)

# Plotting the pie chart
plt.figure(figsize=(10, 6))
plt.pie(df['count'], labels=df['DatabaseDesireNextYear'], autopct='%1.1f%%')
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here

# Query to get WorkWeekHrs and CodeRevHrs for age group 30-35
QUERY = """
SELECT Age, AVG(WorkWeekHrs) as WorkWeekHrs, AVG(CodeRevHrs) as CodeRevHrs
FROM master
WHERE Age BETWEEN 30 AND 35
GROUP BY Age
"""

df = pd.read_sql_query(QUERY, conn)

# Plotting the stacked bar chart
df.set_index('Age').plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Median Work Week Hours and Code Review Hours (Age 30-35)')
plt.xlabel('Age')
plt.ylabel('Hours')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here

# Query to get ConvertedComp for age group 45-60
QUERY = """
SELECT Age, ConvertedComp
FROM master
WHERE Age BETWEEN 45 AND 60
"""

df = pd.read_sql_query(QUERY, conn)

# Group by Age and calculate the median for ConvertedComp
median_df = df.groupby('Age')['ConvertedComp'].median().reset_index()

# Plotting the line chart
plt.figure(figsize=(10, 6))
plt.plot(median_df['Age'], median_df['ConvertedComp'], marker='o')
plt.title('Median Converted Compensation (Age 45-60)')
plt.xlabel('Age')
plt.ylabel('Median Compensation')
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
# Query to get MainBranch data
QUERY = """
SELECT MainBranch, COUNT(*) as count
FROM master
GROUP BY MainBranch
"""

df = pd.read_sql_query(QUERY, conn)

# Plotting the horizontal bar chart
plt.figure(figsize=(10, 6))
df.set_index('MainBranch')['count'].plot(kind='barh', color='green')
plt.title('Main Branch Distribution')
plt.xlabel('Count')
plt.ylabel('Main Branch')
plt.show()

```
